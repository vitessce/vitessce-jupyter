```python
import pandas as pd
```

```python
import altair as alt
```

```python
init_df = pd.read_csv("./o2_sizes.txt", sep="    ", header=None)
```

```python
df = init_df.rename(columns={ 0: "size", 1: "size_human", 2: "filename" })

df["num_cells"] = [1000000, 1000000, 100000, 100000, 10000, 10000, 1000, 1000, 100, 100, 10, 10]
df["num_values"] = df["num_cells"] * 17811
df["format"] = ["csv", "zarr"] * 6
df["size_offset_y"] = [-3000000, 0, 0, 0, 3000000, 500000, 3000000, 500000, 3000000, 500000, 3000000, 500000]
df["size_offset_x"] = [0, 0, 0, 0, 0, 0, 1000, 1000, 100, 100, 30, 30]

df["text_x"] = df["num_cells"] + df["size_offset_x"]
df["text_y"] = df["size"] + (df["size_offset_y"] * 1000)
```

```python
df
```

```python
points = alt.Chart(df).mark_line(point=True).encode(
    x=alt.X("num_cells:Q").scale(type="log").title("Number of cells"),
    y=alt.Y("size:Q").title("Total data size (bytes)"),
    color=alt.Color("format:N").title("Format"),
    shape=alt.Shape("format:N").title("Format"),
    text="size_human:N"
)


text = alt.Chart(df).mark_text(dx = -20, dy=-5).encode(
    x=alt.X("text_x:Q").scale(type="log"),
    y=alt.Y("text_y:Q"),
    color="format:N",
    text="size_human:N"
)


(points + text).properties(title="CSV vs. Zarr file size comparisons")
```
